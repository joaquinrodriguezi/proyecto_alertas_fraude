# Presentación del Proyecto — Sistema de Alertas para Transacciones Sospechosas (ML/Anomalías)

## Introduccion al proyecto
Este proyecto construye un **sistema de alertas** para detectar **transacciones potencialmente sospechosas** a partir de un enfoque de **anomalías** (comportamientos fuera de lo normal) y con **explicaciones claras** de por qué se marca cada caso. La motivación es operativa: cuando el volumen de transacciones crece, aparecen muchos “casos raros” que pueden ser legítimos o señales tempranas de fraude; como no se puede revisar todo, el objetivo es **priorizar** qué revisar primero y con qué evidencias.  
La audiencia principal son **analistas de fraude / riesgo**, equipos de **monitoreo operativo**, y stakeholders de negocio que necesitan **rankings accionables** (Top transacciones, cuentas, comercios y vista por canal) y un puntaje interpretable por transacción. **Importante:** los resultados indican *sospecha*, no fraude confirmado; son una herramienta para priorizar revisión o investigación. 

---

## Preguntas / hipótesis que buscamos responder
1. **¿Qué transacciones deberían revisarse primero?**  
   Hipótesis: las transacciones con **mayor risk_score** y/o múltiples señales relevantes (monto/ratio/velocidad/contexto) concentran la prioridad operativa.

2. **¿Qué “señales explicables” aparecen con más frecuencia en los casos sospechosos?**  
   Hipótesis: ciertas señales de acceso/contexto (ej. *new_device*, *new_ip*, *new_merchant*, *new_location*) se combinan con señales de comportamiento (ej. *high_ratio*, *high_velocity*) en los casos de riesgo alto. 

3. **¿Dónde se concentra el riesgo a nivel de entidades (cuentas y comercios)?**  
   Hipótesis: el riesgo se agrupa en un subconjunto de **cuentas** (múltiples alertas top) y/o **comercios** (concentración de alertas y diversidad de cuentas), útiles para monitoreo y auditoría focalizada.

4. **¿Qué canal presenta mayor exposición relativa al riesgo?**  
   Hipótesis: algunos canales (p.ej. online) muestran mayor proporción de alertas (Top 5%) y mayor score promedio, por lo que conviene priorizar reglas/monitoreo por canal. 

5. **¿Cuánto aportan los outliers “contextuales” (múltiples señales a la vez) a la priorización?**  
   Hipótesis: las transacciones con **2+** y especialmente **3+ señales contextuales** representan un grupo de mayor prioridad para revisión.

6. **¿Qué perfiles de comportamiento diferencian transacciones por canal/tipo?** (EDA bivariado/multivariado)  
   Hipótesis: los montos y la dispersión varían por **Channel** y **TransactionType**, y esas diferencias ayudan a interpretar qué es “anómalo” dentro de cada contexto. 

7. **¿Podemos producir entregables operativos, no solo métricas?**  
   Hipótesis: el pipeline puede generar outputs directamente utilizables (score + motivo por transacción, rankings por entidad/canal y resúmenes) para decidir “por dónde empezar” la revisión diaria.
