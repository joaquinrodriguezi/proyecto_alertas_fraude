# PASO 1: Data Understanding (Entendimiento de Datos)

En este paso realizaremos una exploración inicial del dataset sin realizar transformaciones, limpieza ni imputación. El objetivo es entender la estructura, tipos de datos, calidad y características principales del dataset.

```python
#IMPORTACION DE LIBRERIAS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
# Cargar datos de transacciones
df = pd.read_csv('../data/transacciones_banco.csv')
df.head()
```

## 1. Foto General del Dataset

Verificamos las dimensiones, nombres de columnas y primeros/últimos registros para tener una visión inicial de la estructura.

```python
print("=" * 80)
print("DIMENSIONES DEL DATASET")
print("=" * 80)
print(f"Número de filas: {df.shape[0]:,}")
print(f"Número de columnas: {df.shape[1]}")
print(f"\nTotal de celdas: {df.shape[0] * df.shape[1]:,}")

print("\n" + "=" * 80)
print("NOMBRES DE COLUMNAS")
print("=" * 80)
print(df.columns.tolist())

print("\n" + "=" * 80)
print("PRIMEROS 5 REGISTROS")
print("=" * 80)
display(df.head())

print("\n" + "=" * 80)
print("ÚLTIMOS 5 REGISTROS")
print("=" * 80)
display(df.tail())
```

## 2. Tipos de Datos

Analizamos los tipos de datos de cada columna para entender cómo están almacenados (numéricos, objetos, fechas, etc.).

```python
print("=" * 80)
print("TIPOS DE DATOS POR COLUMNA")
print("=" * 80)
print(df.dtypes)

print("\n" + "=" * 80)
print("INFORMACIÓN GENERAL DEL DATASET")
print("=" * 80)
df.info()
```

## 3. Calidad Superficial de Datos

Revisamos la presencia de valores nulos y registros duplicados para evaluar la calidad inicial del dataset.

```python
print("=" * 80)
print("VALORES NULOS POR COLUMNA")
print("=" * 80)
nulos = df.isnull().sum()
nulos_pct = (df.isnull().sum() / len(df)) * 100
resumen_nulos = pd.DataFrame({
    'Nulos': nulos,
    'Porcentaje': nulos_pct
})
print(resumen_nulos[resumen_nulos['Nulos'] > 0])

print(f"\nTotal de valores nulos en el dataset: {df.isnull().sum().sum():,}")
print(f"Columnas con valores nulos: {(nulos > 0).sum()} de {len(df.columns)}")

print("\n" + "=" * 80)
print("REGISTROS DUPLICADOS")
print("=" * 80)
duplicados = df.duplicated().sum()
print(f"Número de registros duplicados: {duplicados:,}")
print(f"Porcentaje de duplicados: {(duplicados / len(df)) * 100:.2f}%")
```

## 4. Resumen Estadístico de Variables Numéricas

Analizamos las estadísticas descriptivas (media, mediana, desviación estándar, min, max, cuartiles) de las variables numéricas clave.

```python
print("=" * 80)
print("RESUMEN ESTADÍSTICO DE VARIABLES NUMÉRICAS")
print("=" * 80)

# Columnas numéricas de interés
columnas_numericas = ['TransactionAmount', 'AccountBalance', 'CustomerAge', 
                      'TransactionDuration', 'LoginAttempts']

# Verificar cuáles existen en el dataset
columnas_existentes = [col for col in columnas_numericas if col in df.columns]

if columnas_existentes:
    display(df[columnas_existentes].describe())
else:
    print("No se encontraron las columnas numéricas esperadas")
    print("Columnas numéricas disponibles:")
    print(df.select_dtypes(include=[np.number]).columns.tolist())
```

## 5. Distribución de Frecuencias de Variables Categóricas

Exploramos las frecuencias y distribuciones de las variables categóricas principales para entender los valores más comunes.

```python
print("=" * 80)
print("FRECUENCIAS DE VARIABLES CATEGÓRICAS")
print("=" * 80)

# TransactionType
if 'TransactionType' in df.columns:
    print("\nTransactionType:")
    print(df['TransactionType'].value_counts())
    print(f"Total categorías: {df['TransactionType'].nunique()}")

# Channel
if 'Channel' in df.columns:
    print("\n" + "-" * 80)
    print("\nChannel:")
    print(df['Channel'].value_counts())
    print(f"Total categorías: {df['Channel'].nunique()}")

# CustomerOccupation
if 'CustomerOccupation' in df.columns:
    print("\n" + "-" * 80)
    print("\nCustomerOccupation:")
    print(df['CustomerOccupation'].value_counts())
    print(f"Total categorías: {df['CustomerOccupation'].nunique()}")

# Location (Top 10)
if 'Location' in df.columns:
    print("\n" + "-" * 80)
    print("\nLocation (Top 10):")
    print(df['Location'].value_counts().head(10))
    print(f"Total ubicaciones únicas: {df['Location'].nunique()}")

# MerchantID (Top 10)
if 'MerchantID' in df.columns:
    print("\n" + "-" * 80)
    print("\nMerchantID (Top 10):")
    print(df['MerchantID'].value_counts().head(10))
    print(f"Total comercios únicos: {df['MerchantID'].nunique()}")
```

## 6. Cardinalidad de Identificadores

Evaluamos la cardinalidad (número de valores únicos) de los campos identificadores para verificar su unicidad y detectar posibles problemas.

```python
print("=" * 80)
print("CARDINALIDAD DE IDENTIFICADORES")
print("=" * 80)

# Columnas identificadoras
columnas_ids = ['TransactionID', 'AccountID', 'DeviceID', 'IP Address', 
                'MerchantID', 'Location']

cardinalidad = {}
for col in columnas_ids:
    if col in df.columns:
        valores_unicos = df[col].nunique()
        total_valores = len(df)
        porcentaje = (valores_unicos / total_valores) * 100
        cardinalidad[col] = {
            'Valores únicos': valores_unicos,
            'Total registros': total_valores,
            'Porcentaje unicidad': f"{porcentaje:.2f}%"
        }
        
resumen_cardinalidad = pd.DataFrame(cardinalidad).T
display(resumen_cardinalidad)

print("\nInterpretación:")
print("- 100% = Cada valor es único (clave primaria)")
print("- <100% = Valores repetidos (pueden ser válidos según el campo)")
```

## 7. Análisis Temporal

Parseamos las columnas de fechas a formato datetime y analizamos los rangos temporales para entender el periodo cubierto por los datos.

```python
print("=" * 80)
print("ANÁLISIS TEMPORAL")
print("=" * 80)

# TransactionDate
if 'TransactionDate' in df.columns:
    print("\nTransactionDate:")
    print(f"  Tipo de dato original: {df['TransactionDate'].dtype}")
    print(f"  Muestra de valores: {df['TransactionDate'].head(3).tolist()}")
    
    # Parsear a datetime
    transaction_dates = pd.to_datetime(df['TransactionDate'], errors='coerce')
    print(f"\n  Rango temporal:")
    print(f"    - Fecha mínima: {transaction_dates.min()}")
    print(f"    - Fecha máxima: {transaction_dates.max()}")
    print(f"    - Duración: {(transaction_dates.max() - transaction_dates.min()).days} días")
    print(f"  Valores no parseables: {transaction_dates.isna().sum()} ({(transaction_dates.isna().sum()/len(df)*100):.2f}%)")

# PreviousTransactionDate
if 'PreviousTransactionDate' in df.columns:
    print("\n" + "-" * 80)
    print("\nPreviousTransactionDate:")
    print(f"  Tipo de dato original: {df['PreviousTransactionDate'].dtype}")
    print(f"  Muestra de valores: {df['PreviousTransactionDate'].head(3).tolist()}")
    
    # Parsear a datetime
    prev_dates = pd.to_datetime(df['PreviousTransactionDate'], errors='coerce')
    print(f"\n  Rango temporal:")
    print(f"    - Fecha mínima: {prev_dates.min()}")
    print(f"    - Fecha máxima: {prev_dates.max()}")
    if pd.notna(prev_dates.min()) and pd.notna(prev_dates.max()):
        print(f"    - Duración: {(prev_dates.max() - prev_dates.min()).days} días")
    print(f"  Valores no parseables: {prev_dates.isna().sum()} ({(prev_dates.isna().sum()/len(df)*100):.2f}%)")
```
